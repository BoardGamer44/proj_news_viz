## Протип

Для протипа возьмем 100 новостей из датасета. На этой выборке нужно понять какие графики будем использовать, которые наиболее точно будут отражать наличие топонимов в датасете. А также выявить возможные проблемы

```python
import pandas as pd
```

```python
dateparse = lambda x: pd.datetime.strptime(x, "%Y-%m-%d %H:%M:%S")
```

```python
!ls
```

```python
data = pd.read_csv('rt_test_geo.csv',
                      sep=';',
                      index_col=False,
                      escapechar='\\',
                      parse_dates=['date'],
                      date_parser=dateparse,
                      quotechar='\"')
```

```python
data.head()
```

```python
def show_entities(text):
    lst = Counter()
    try:
        result = processor(text)
    except(KeyError):
        return None
    for item in result.matches:
        #print(item.referent.slots)
        value = getattr(item.referent, 'name')
        lst[value] += 1
    return lst
```

```python
print(show_entities('В России много друей'))
```

```python
data['entities'] = data['text'].apply(show_entities)
```

```python
data.head()
```

```python
print('Кол-во строк с значением NaN в столбце "entities" = ', data['entities'].isnull().sum())
data[data['entities'].isnull()]
```

Соберем получившиеся данные в одну структуру

```python
full_counter = Counter()
for i in data.entities.values:
    if i != None:
        full_counter += i
```

```python
full_counter.most_common()
```

```python
import matplotlib.pyplot as plt
```

```python
x = dict(full_counter.most_common(20)).keys()
y = dict(full_counter.most_common(20)).values()

fig, ax = plt.subplots()

ax.bar(x, y)
ax.set_title('Топонимы')
ax.set_ylabel('Частота')
ax.set_xlabel('Топоним')

ax.set_facecolor('seashell')
fig.set_facecolor('floralwhite')
fig.set_figwidth(40)    #  ширина Figure
fig.set_figheight(15)    #  высота Figure


plt.show()
```
